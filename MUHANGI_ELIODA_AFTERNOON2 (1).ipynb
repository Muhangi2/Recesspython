```python
print("\nMissing Values:")
df.isnull().sum()
```

```python
df.dropna(subset=["Duration", "Pulse", "Maxpulse", "Calories"])
```

```python
df["Calories per Minute"] = df["Calories"] / df["Duration"]
 df["Calories per Minute"].mean()
```

```python
df["Calories per Minute"] = df["Calories"] / df["Duration"]
average_calories_per_minute = df["Calories per Minute"].mean()
print("\n Average calories in every minute:", average_calories_per_minute)
```

```python
z_scores = (df["Pulse"] - df["Pulse"].mean()) / df["Pulse"].std()
df = df[(z_scores < 3)]
print("\nAfter removing outliers using z-score:")
print(df)
```

```python
print("\nAfter removing duplicates based on 'Duration' and 'Pulse' columns:")
df = df.drop_duplicates(subset=["Duration", "Pulse"])
df
```

```python
df = df.fillna(df.mode().iloc[0])
print("\n   After filling missing values with mode  ")
df
```

```python
missing_values = df.isnull().sum()
print("\n  Missing Values:")
print(missing_values)
```
